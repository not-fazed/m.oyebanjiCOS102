```python
import tkinter as tk

class Zenith():
    def unique_services(self):
        pass
    def mutual_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"
        ] 
class GlobalBanking(Zenith):
    def unique_services(self):
        return [
            "Multi-currency management",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]
class CommercialBanking(Zenith):
    def unique_services(self):
        return [
            "Advisory services"
        ]                   
class RetailBanking(Zenith):
    def unique_services(self):
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]


def show_services():
    name = name_entry.get()
    division = division_entry.get().lower()

    if division == "retail":
        obj = RetailBanking()
    elif division == "global":
        obj = GlobalBanking()
    elif division == "commercial":
        obj = CommercialBanking()
    else:
        result_label.config(text="Invalid division")
        return

    services = obj.mutual_services() + obj.unique_services()

    result_label.config(
        text=f"{name}'s Services:\n" + "\n".join(services)
    )


root = tk.Tk()
root.title("Zenith Bank Services")

tk.Label(root, text="Employee Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Division (Retail/Global/Commercial)").pack()
division_entry = tk.Entry(root)
division_entry.pack()

tk.Button(root, text="Show Services", command=show_services).pack()

result_label = tk.Label(root, text="")
result_label.pack()

root.mainloop()
```
