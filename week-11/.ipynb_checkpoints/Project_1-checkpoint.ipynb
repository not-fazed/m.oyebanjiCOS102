```python
from abc import ABC, abstractmethod

# 1. Define the Abstract Base Class
class coup_de_ecriva(ABC):
    
    @abstractmethod
    def Fan_Page(self):
        pass

# 2. Define the Subclasses using Single Inheritance
class FC_Cirok(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to FC Cirok")

class Madiba_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Madiba FC")

class Blue_Jay_FC(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to Blue Jay FC")

class TSG_Walker(coup_de_ecriva):
    def Fan_Page(self):
        print("Welcome to TSG Walker")

# 3. Main Program Logic for User Input and Instantiation
def main():
    print("--- Welcome to the coup_de_ecriva Portal ---")
    print("Available clubs: FC Cirok, Madiba FC, Blue Jay FC, TSG Walker")
    
    # Receive input from the user
    user_choice = input("Enter the name of the club you support: ").strip()
    
    # Standardize input to map to the correct class instantiation
    # (Using lower case comparison to make it more user-friendly and case-insensitive)
    choice_lower = user_choice.lower()
    
    club_object = None
    
    if choice_lower in ["fc cirok", "cirok"]:
        club_object = FC_Cirok()
    elif choice_lower in ["madiba fc", "madiba"]:
        club_object = Madiba_FC()
    elif choice_lower in ["blue jay fc", "blue jay"]:
        club_object = Blue_Jay_FC()
    elif choice_lower in ["tsg walker", "walker"]:
        club_object = TSG_Walker()
    else:
        print(f"Sorry, '{user_choice}' is not a recognized club in this system.")
        return

    # Invoke the abstract method implementation from the subclass
    club_object.Fan_Page()

if __name__ == "__main__":
    main()
```
