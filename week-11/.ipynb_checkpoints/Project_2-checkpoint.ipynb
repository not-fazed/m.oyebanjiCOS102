```python
import tkinter as tk
from tkinter import ttk, messagebox
from abc import ABC, abstractmethod

class External_Vendors(ABC):
    
    @abstractmethod
    def menu(self):
        """Abstract method to return the vendor's menu details."""
        pass

class Cooperative_Cafeteria(External_Vendors):
    def menu(self):
        return [
            ("Jollof Rice and Stew", 200),
            ("White Rice and Stew", 200),
            ("Fried Rice", 200),
            ("Salad", 100),
            ("Plantain", 100)
        ]

class Faith_Hostel_Cafeteria(External_Vendors):
    def menu(self):
        return [
            ("Fried Rice", 400),
            ("White Rice and Stew", 400),
            ("Jollof Rice", 400),
            ("Beans", 200),
            ("Chicken", 1000)
        ]

class Student_Centre_Cafeteria(External_Vendors):
    def menu(self):
        return [
            ("Chicken Fried Rice", 800),
            ("Pomo Sauce", 300),
            ("Spaghetti Jollof", 500),
            ("Amala/Ewedu", 500),
            ("Semo with Eforiro Soup", 500)
        ]
class VendorManagementApp:
    def __init__(self, root):
        self.root = root
        self.root.title("PAU External Food Vendors Management System")
        self.root.geometry("550x450")
        self.root.resizable(False, False)
        
        # Style layout configurations
        self.style = ttk.Style()
        self.style.configure("TLabel", font=("Helvetica", 11))
        self.style.configure("TButton", font=("Helvetica", 10, "bold"))
        
        # Header Label
        header = tk.Label(root, text="PAU External Food Vendors", font=("Helvetica", 16, "bold"), bg="#800020", fg="white", pady=10)
        header.pack(fill=tk.X)
        
        # Selection Frame
        frame = ttk.Frame(root, padding="20 20 20 10")
        frame.pack(fill=tk.X)
        
        ttk.Label(frame, text="Select Vendor of Interest:", font=("Helvetica", 12, "bold")).grid(row=0, column=0, sticky=tk.W, padx=5)
        
        # Dropdown selection for Vendors
        self.vendor_options = ["Cooperative Cafeteria", "Faith Hostel Cafeteria", "Student Centre Cafeteria"]
        self.selected_vendor = tk.StringVar()
        self.vendor_dropdown = ttk.Combobox(frame, textvariable=self.selected_vendor, values=self.vendor_options, state="readonly", width=25, font=("Helvetica", 11))
        self.vendor_dropdown.grid(row=0, column=1, padx=10)
        self.vendor_dropdown.current(0) # Default to first choice
        
        # View Menu Button
        view_btn = ttk.Button(frame, text="Display Menu", command=self.display_vendor_menu)
        view_btn.grid(row=0, column=2)
```
