```python
#ATR calculator for izfin technology
print("Hello! and welcome to izfin technology")
print("All fields must be filled with years")
xp = int(input("What is your work experience"))
age = int(input("How old are you?"))
if xp > 25 and age > 55 :
    print("Your ATR is N5600000")
elif xp > 20 and age >= 45 :
    print("Your ATR IS N448000")
elif xp > 10 and age >= 35 :
    print("Your ATR is N1500000")
elif xp < 10 and age < 35 :
    print("Your ATR is N550000")
```
