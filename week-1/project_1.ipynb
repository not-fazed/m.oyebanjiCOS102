```python
# prgram to calculate these python program 
print("Select the problem you want to solve ")
print("1. Simple Interest")
print("2. Compound Interest")
print("3. Annuity Plan ")
      
problem = input("Input a number ")
if problem == "1":
    P = float(input("Input P"))
    R = float(input("Input R"))
    T = float(input("Input T"))
    A = P*(1+(R/100 )*T)
    print("Your SIMPLE INTEREST in ",A)
if problem == "2":
    P = float(input("Input P"))
    R = float(input("Input R"))
    T = float(input("Input T"))
    N = int(input("Input N"))
    CA = P*(1+(R/N)**T)
    print("Your COMPOUND INTEREST in ",CA)
if problem == "3":
    P = float(input("Input P"))
    M = float(input("Input M"))
    R = float(input("Input R"))
    T = float(input("Input T"))
    N = int(input("Input N"))
    AP = P*M*T * (((1+(R/N))**T)-1) / (R/N)
    print("Your ANNUITY PLAN in ",AP)

if problem >= "4":
    print("NOT A VALID OPTION")
```
