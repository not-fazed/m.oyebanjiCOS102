```python
#Employee login app
import panda as pd
data = pd.read_csv("GIG-logistics.csv")
def check_employee(name, department):
    return data[
        (data["Name"].str.lower() == name.lower()) &
        (data["Department"].str.lower() == department.lower())
    ]
def verify_employee():
    name = name_entry.get()
    department = dept_entry.get()
    
    result = check_employee(name, department)
    
    if not result.empty:
        employees_in_dept = data[
            data["Department"].str.lower() == department.lower()
        ]["Name"].tolist()
        
        message = f"Welcome {name}!\n\nOther members in {department}:\n"
        message += "\n".join(employees_in_dept)
        
        messagebox.showinfo("Success", message)
    else:
        messagebox.showerror("Error", "Employee does not exist")
root = tk.Tk()
root.title("GIG Logistics Employee Checker")
root.geometry("400x300")

tk.Label(root, text="Enter Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Check Employee", command=verify_employee).pack()

root.mainloop()

```
