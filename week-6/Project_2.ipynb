```python
import tkinter as tk

class Shipping:
    def __init__(self, destination, mass):
        self.destination = destination.strip().lower()
        self.mass = mass

    def get_price(self):
        if self.destination == "pau":
            return 2000 if self.mass >= 10 else 1500
        elif self.destination == "epe":
            return 5000 if self.mass >= 10 else 4000
        else:
            return None


# Button Logic
def compute_fee():
    dest = dest_entry.get()

    try:
        mass = float(mass_entry.get())
    except ValueError:
        output_label.config(text="Enter a valid number for weight.")
        return

    # Validation for zero or negative weight
    if mass <= 0:
        output_label.config(text="Weight must be greater than 0.")
        return

    order = Shipping(dest, mass)
    price = order.get_price()

    if price is None:
        output_label.config(text="Location not supported (PAU or Epe only).")
    else:
        output_label.config(text=f"Fee: ₦{price}")


# GUI Setup
app = tk.Tk()
app.title("Shipping Cost Checker")
app.geometry("400x300")

# UI Elements
tk.Label(app, text="Destination (PAU or Epe):").pack(pady=10)
dest_entry = tk.Entry(app)
dest_entry.pack()

tk.Label(app, text="Weight (kg):").pack(pady=10)
mass_entry = tk.Entry(app)
mass_entry.pack()

tk.Button(app, text="CHECK PRICE", command=compute_fee).pack(pady=15)

output_label = tk.Label(app, text="")
output_label.pack(pady=20)

# Run
app.mainloop()
```
